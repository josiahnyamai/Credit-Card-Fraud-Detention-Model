```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier


from sklearn.metrics  import accuracy_score

```

**Load data**

```python
data = pd.read_csv('creditcard.csv')
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Class'].value_counts()
```

**The dataset is highly unbalanced**

**0 -> Normal**

**1 -> Fraud**

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Statistical Measures of the data**

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

**Compare values for both transactions**

```python
data.groupby('Class').mean()
```

**Build a Sample dataset contaning similar distribution of normal and Fraudlent transactions**

```python
legit_sample_data = legit.sample(n=492)
```

**Concatenating two DataFrames(legit_sample_data and Fraud datasets)**

```python
new_data = pd.concat([legit_sample_data, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
new_data.to_csv("creditcardfrauddata.csv",index=False)
```

**Spliting the data**

```python
x = new_data.drop(columns='Class', axis=1)
y = new_data['Class']
```

```python
x
```

```python
y
```

**Split the data into training and testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

**Logistic Regression Model Training**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

**Accuracy on training data**

```python
X_train_prediction = model.predict(X_train)
accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on Training data : ", accuracy)
```

**Accuracy on testing data**

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy on Testing data : ", test_accuracy)
```

**Decision Tree Model Training**

```python
modelD = DecisionTreeClassifier()
```

```python
modelD.fit(X_train, Y_train)
```

**Model Evaluation**

**Accuracy on Training data**

```python
train_prediction = modelD.predict(X_train)
t_accuracy = accuracy_score(train_prediction, Y_train)
print("Decision Tree accuracy on training data ", t_accuracy)
```

**Accuracy on Testing data**

```python
test_prediction = modelD.predict(X_test)
te_accuracy = accuracy_score(test_prediction, Y_test)
print("Decision Tree accuracy on testing data ", te_accuracy)
```

**Serialization**

```python
import pickle
pickle_out = open("model.pkl", "wb")
pickle.dump(model,pickle_out)
pickle_out.close()
```
